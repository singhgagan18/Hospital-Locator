# Running K-Means Algorithm

```python
from sklearn.cluster import KMeans
K_clusters = range(1,10)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
Y_axis = df[['Latitude']]
X_axis = df[['Longitude']]
score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]
# Visualize
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python

kmeans = KMeans(n_clusters = 3, init ='k-means++')
kmeans.fit(df[df.columns[3:5]]) # Compute k-means clustering.
df['cluster_label'] = kmeans.fit_predict(df[df.columns[3:5]])
centers = kmeans.cluster_centers_ # Coordinates of cluster centers.
labels = kmeans.predict(df[df.columns[3:5]]) # Labels of each point
df.head(10)
```

```python
#plotting clusters after running K-means
df.plot.scatter(x = 'Latitude', y = 'Longitude', c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
#creating functions to create map as per the user input
cols = ['#e6194b','#3cb44b','#ffe119','#4363d8']

def checkcity(cityval):
    
    for index, row in df.iterrows():
        if row['City'] == cityval:
            x = row['Latitude']
            y = row['Longitude']
    m = create_map(df, 'cluster_label',cityval,x,y)
    return m

def create_map(df, cluster_label,cityval,x,y):
    m = folium.Map(location=[x, y], zoom_start = 9, tiles = 'Stamen Toner')
  
  
    for _, row in df.iterrows():
    
        if row.City == cityval:
        
            color_dict = {0: 'lightblue', 1: 'lightgreen', 2: 'orange', 3: 'red'}

            mystring = 'Name of the Hospital:{}, Type of the Hospital: {}, Available bed in Hospital:{}'.format(row.Hospital_Name, row.Htype, row.Beds)

            folium.CircleMarker(
              location = [row.Latitude,row.Longitude],
              radius = 5,
              popup = mystring,
              columns = ['City', 'Zip_code'],
              colors = color_dict,
              tooltip='Click here to see Popup',
              fill =True,
              key_on=df.City,
              fill_colors= color_dict,
              legend_name = 'Number of Hospital per City'
      
            ).add_to(m)
  
    return m
    






```

```python
#checking the best clusters and silhoutte score
from tqdm import tqdm
from sklearn.metrics import silhouette_score
best_silhouette, best_k = 0,1

for k in tqdm(range(2,100)):
  model = KMeans(n_clusters=k, random_state=1).fit(X)
  class_predictions = model.predict(X)


  curr_silhouette = silhouette_score(X, class_predictions)
  if curr_silhouette>best_silhouette:
    best_k = k
    best_silhouette = curr_silhouette

print(f'k = {best_k}')
print(f'Silhouette Score: {best_silhouette}')
```
